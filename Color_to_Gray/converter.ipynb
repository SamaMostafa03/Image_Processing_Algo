Input Image:

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

input_image = Image.open("../input_images/flower.jpg")
image_array = np.array(input_image)

print(f"Shape: {image_array.shape}")  #(height, width, color channels)
```

Save and Plot the results:

```python
def show_results(output_array,output_image_name,input_image_array=input_image):
    Image.fromarray(output_array.astype(np.uint8)).save("../output_images/"+output_image_name+"_image.bmp")
    plt.figure(figsize=(10, 5))
    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(np.array(input_image_array), cmap="gray")
    plt.title("Original Image", fontsize=16)
    plt.axis("off")
    # Output Image
    plt.subplot(1, 2, 2)
    plt.imshow(output_array, cmap="gray")
    plt.title(output_image_name+" Edge Detector", fontsize=16)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```

Convert Color Image to Grayscale:

```python
def convert_to_grayscale(image_array):
    height, width, _ = image_array.shape
    grayscale_array = np.zeros((height, width), dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            r, g, b = image_array[i, j]
            gray_value = int(0.299 * r + 0.587 * g + 0.114 * b)
            grayscale_array[i, j] = gray_value
    return grayscale_array

show_results(output_array=convert_to_grayscale(image_array), output_image_name="gray_scale")
```
