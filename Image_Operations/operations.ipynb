Input Image:

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

input_image = Image.open("../input_images/WholeHouse.tif")
image_array = np.array(input_image)
height, width = image_array.shape

print(image_array)
print(f"Shape: {image_array.shape}")  #(height, width)
```

Save and Plot the results:

```python
def show_results(output_array,output_image_name,input_image_array=input_image):
    Image.fromarray(output_array.astype(np.uint8)).save("../output_images/"+output_image_name+"_image.bmp")
    plt.figure(figsize=(10, 5))
    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(np.array(input_image_array), cmap="gray")
    plt.title("Original Image", fontsize=16)
    plt.axis("off")
    # detector Image
    plt.subplot(1, 2, 2)
    plt.imshow(output_array, cmap="gray")
    plt.title(output_image_name+" Image", fontsize=16)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```

Image Addition Algorithm:

```python
def add_image(image_array1,image_array2):
    output_array= np.zeros_like(image_array1,dtype=np.uint8)
    image_array1 = image_array1.astype(np.int16)
    image_array2 = image_array2.astype(np.int16)
    output_array= image_array1+image_array2
    output_array=np.clip(output_array, 0, 255)
    return output_array

show_results(output_array=add_image(image_array, image_array), output_image_name="Addition")
```

Image Substraction Algorithm:

```python
def sub_image(image_array1,image_array2):
    output_array= np.zeros_like(image_array1)
    output_array= image_array1-image_array2
    output_array=np.clip(output_array, 0, 255)
    return output_array

show_results(output_array=sub_image(image_array, image_array), output_image_name="Substration")
```

Image Invertion Algorithm:

```python
def invert(image_array):
    output_array= np.zeros_like(image_array)
    output_array= 255-image_array
    return output_array

show_results(output_array=invert(image_array), output_image_name="Invertion")
```
